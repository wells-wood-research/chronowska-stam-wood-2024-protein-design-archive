```python
data = pd.read_json("./data/20240827_data.json")
pdb_release = pd.read_csv("./data/pdb_release_dates/combined.csv")
```

# MMseqs2

```python
mm_chains_dvd = pd.read_csv("./similarity/mmseq/try5/d-v-d/chains/results.m8", sep="\t", header=None, index_col=False, names=["query_name", "query_length", "target_name", "target_length", "bits", "pident", "cigar", "fragment_of_query_covered_by_target", "fragment_of_target_covered_by_query"])
mm_struct_dvd = pd.read_csv("./similarity/mmseq/try5/d-v-d/struct/results.m8", sep="\t", header=None, index_col=False, names=["query_name", "query_length", "target_name", "target_length", "bits", "pident", "cigar", "fragment_of_query_covered_by_target", "fragment_of_target_covered_by_query"])
mm_chains_dvp = pd.read_csv("./similarity/mmseq/try4/d-v-p/chains/results.m8", sep="\t", header=None, index_col=False, names=["query_name", "query_length", "target_name", "target_length", "bits", "pident", "cigar", "fragment_of_query_covered_by_target", "fragment_of_target_covered_by_query"])
mm_struct_dvp = pd.read_csv("./similarity/mmseq/try4/d-v-p/struct/results.m8", sep="\t", header=None, index_col=False, names=["query_name", "query_length", "target_name", "target_length", "bits", "pident", "cigar", "fragment_of_query_covered_by_target", "fragment_of_target_covered_by_query"])
```

```python
data_dict = dict(zip(data["pdb"].str.lower(), data["release_date"]))
pdb_release_dict = dict(zip(pdb_release["Entry ID"].str.lower(), pdb_release["Release Date"]))
excluded_pdbs = set(data["pdb"].str.lower())
```

```python
seen = set()
mm_chain_dvd_highest_bits_dict = {}
mm_struct_dvd_highest_bits_dict = {}
mm_chain_dvp_highest_pident_dict = {}
mm_struct_dvp_highest_pident_dict = {}
```

```python
file_name = mm_chains_dvd
dict_name = mm_chain_dvd_highest_bits_dict

for i, row in file_name.iterrows():
    if i % 1000 == 0:
        print(f"{i}/{len(file_name)}")
        
    query_name = row["query_name"][:4].lower()
    target_name = row["target_name"][:4].lower()
    
    if query_name == target_name:
        continue

    try:
        query_release = datetime.date.fromisoformat(data_dict[query_name])
        target_release = datetime.date.fromisoformat(pdb_release_dict.get(target_name, "1900-01-01"))
    except KeyError as e:
        print(f"Error at query: {query_name}. Missing release date for target: {target_name}")
        continue

    if query_release < target_release:
        continue
    
    similarity = row["bits"]
    if query_name not in dict_name or dict_name[query_name]["sim"] < similarity:
        dict_name[query_name] = {"sim": similarity, "partner": target_name}
```

```python
with open("./similarity/mmseq/try5/d-v-d/chains/mm_bits_chains_alone.json", "w") as outfile: 
    json.dump(dict_name, outfile)
```

```python
# For whole structures
file_name = mm_struct_dvd
dict_name = mm_struct_dvd_highest_bits_dict

for i, row in file_name.iterrows():
    if i % 1000 == 0:
        print(f"{i}/{len(file_name)}")
        
    query_name = row["query_name"][:4].lower()
    target_name = row["target_name"][:4].lower()
    
    if query_name == target_name:
        continue

    try:
        query_release = datetime.date.fromisoformat(data_dict[query_name])
        target_release = datetime.date.fromisoformat(pdb_release_dict.get(target_name, "1900-01-01"))
    except KeyError as e:
        print(f"Error at query: {query_name}. Missing release date for target: {target_name}")
        continue

    if query_release < target_release:
        continue
    
    similarity = row["bits"]
    if query_name not in dict_name or dict_name[query_name]["sim"] < similarity:
        dict_name[query_name] = {"sim": similarity, "partner": target_name}
```

```python
with open("./similarity/mmseq/try5/d-v-d/struct/mm_bits_chains_alone.json", "w") as outfile: 
    json.dump(dict_name, outfile)
```

```python
with open("./similarity/mmseq/try4/d-v-p/chains/mm_pident_chains_alone.json", "w") as outfile: 
    json.dump(dict_name, outfile)
```

```python
# For whole structures
file_name = mm_struct_dvp
dict_name = mm_struct_dvp_highest_pident_dict

for i, row in file_name.iterrows():
    if i % 1000 == 0:
        print(f"{i}/{len(file_name)}")
        
    query_name = row["query_name"][:4].lower()
    target_name = row["target_name"][:4].lower()
    
    if query_name == target_name or target_name in excluded_pdbs:
        continue

    try:
        query_release = datetime.date.fromisoformat(data_dict[query_name])
        target_release = datetime.date.fromisoformat(pdb_release_dict.get(target_name, "1900-01-01"))
    except KeyError as e:
        print(f"Error at query: {query_name}. Missing release date for target: {target_name}")
        continue

    if query_release < target_release:
        continue
    
    similarity = row["pident"]
    if query_name not in dict_name or dict_name[query_name]["sim"] < similarity:
        dict_name[query_name] = {"sim": similarity, "partner": target_name}
```
